```python
target_names = dict(enumerate(iris.target_names))
```

```python
print(target_names)
```

```python
df["Species"] = df["target"].map(target_names)
```

```python
#Quick Pandas inspection
print("\nShape:",df.shape)
print("-"*50)
print("\nInfo:")
print(df.info())
print("\nDescribe (numeric):")
print(df.describe())
```

```python
# convert to numpy arrays 
feature_cols = iris.feature_names
X = df[feature_cols]
Y = df["target"]
```

```python
print(type(X))
```

```python
print(type(Y))
```

```python
print(Y)
```

```python
print(X)
```

```python
X = X.to_numpy()
Y = Y.to_numpy()
```

```python
# Numpy stats(column-wise):mean,std
col_means = X.mean(axis=1)
col_stds = X.std(axis=0,ddof=0)
print("\n Numpy column means :",col_means)
print("\n Numpy column stds :",col_stds)
```

```python
print("Missing Values in the dataframe")
print(df.isnull().sum())
```

```python
df.groupby("Species")[feature_cols].mean()
```

```python
# pair plot with seaborn
sns.set_style("whitegrid")

sns.pairplot(df[feature_cols + ["Species"]] , hue="Species",diag_kind="kde")
plt.suptitle("Pairwise Feature Realtionship",y=1.02)
plt.show()
```

```python
corr = df[feature_cols].corr(numeric_only=True)
plt.figure(figsize=(6,4))
sns.heatmap(corr,annot=True,cmap="coolwarm",linewidths=0.5)
plt.title("Corelation - heatmap")
plt.show()
```

```python
# Class distribution bar chart 
plt.figure(figsize=(6,4))
sns.countplot(x="Species",data=df)
plt.title("Class distribution")
plt.xlabel("Species")
plt.ylabel("Count")

plt.tight_layout()
plt.show()
```

```python
# Simple scatter with hue (example: petal length vs petal width)
plt.figure(figsize=(6,4))
sns.scatterplot(
    data = df,
    x="petal length (cm)",
    y="petal width (cm)",
    hue="Species"
)
plt.title("Petal Length VS Petal Width by Species")
plt.tight_layout()
plt.show()
```
